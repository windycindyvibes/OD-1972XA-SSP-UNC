```python
import rebound
from rebound import hash as h
```

```python
sa = rebound.Simulationarchive("archive_new.bin")
sim = sa[-1]
```

```python
archive = "archive_new_1.bin"
tstart = sim.t
tout = 1000
tend = 50e6

sa = sim.save_to_file(archive, interval=tout, delete_file=True)
```

```python
sim.collision_resolve = "halt"
```

```python
N_pl = 8
```

```python
import numpy as np
times = np.arange(tstart, tend, tout)
Nsteps = len(times)

for i in range(Nsteps):
    sim.integrate(times[i], exact_finish_time=0)
    print("Time {0:7.4f} Myr --Fraction Done {1:5.4} -- # of clones {2}".format(sim.t/1e6, (sim.t-tstart)/(tend-tstart), sim.N-N_pl))
```
