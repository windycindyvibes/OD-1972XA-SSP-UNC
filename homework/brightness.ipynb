```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
A_1 = 3.33
B_1 = 0.63
A_2 = 1.87
B_2 = 1.22
G = 0.15
H = 14
```

```python
def phi_1(alpha):
    return math.e**(-A_1*(np.tan((alpha/2))**B_1))

def phi_2(alpha):
    return math.e**(-A_2*(np.tan((alpha/2))**B_2))

def get_H(alpha):
    print(H)
    alpha = [H - 2.5*math.log10((1-G)*phi_1(value)+G*phi_2(value)) for value in alpha]
    return alpha
```

```python
x = np.linspace(0, 30, 400)
x_rad = np.radians(x)
y = get_H(x_rad)

plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title('Degree vs. Reduced V-mag')
plt.xlabel('Degree')
plt.ylabel('Reduced V-mag')
plt.grid(True)
plt.legend()
plt.show()
```
