```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 24, 400)
lat = np.radians(35+54/60+50.58/3600)
dec = np.radians(10+13/60+49/3600)
y = np.arcsin(np.sin(lat)*np.sin(dec) + np.cos(lat)*np.cos(dec)*np.cos(np.radians(15*x)))
y = np.degrees(y)

plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title('Altitude vs. Hour Angle')
plt.xlabel('Hour Angle')
plt.ylabel('Altitude')
plt.grid(True)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 24, 400)
lat = np.radians(35+54/60+50.58/3600)
dec = np.radians(10+13/60+49/3600)
h = np.arcsin(np.sin(lat)*np.sin(dec) + np.cos(lat)*np.cos(dec)*np.cos(np.radians(15*x)))
h = np.degrees(h)

y = np.arcsin((-1*np.cos(dec)*np.sin(np.radians(x)))/np.cos(np.radians(h)))
y = np.degrees(y)

plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title('Azimuth vs. Hour Angle')
plt.xlabel('Hour Angle')
plt.ylabel('Azimuth')
plt.grid(True)
plt.legend()
plt.show()
```
